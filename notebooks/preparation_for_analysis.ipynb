```python
import matplotlib.pyplot as plt
import pandas as pd
import codecs
import json
```

```python
data_dir = "../experimental_data"
questionnaire_results_file = data_dir + "/data.json"
json_data = json.load(codecs.open(questionnaire_results_file, 'r','utf-8'))
result_dict ={}
for i, json_dict in enumerate(json_data[9 : 49]):
    image_name = json_dict["imgSrc"] 
    answer = json_dict["answer"]
    result_dict[image_name] = answer
results = []
trial_count = 1
for image_name, answers in result_dict.items():
    results.append([trial_count, *answers, image_name])
    trial_count += 1
columns=["trial_count", "answer(q1)", "answer(q2)", "answer(q3)", "answer(q4)", "answer(q5)", "answer(q6)", "image_name"]
result_df = pd.DataFrame(results, columns=columns)
# display(result_df)
LF_HF_file = data_dir + "/20191212_LFHF.txt"
LFHF_data = pd.read_csv(LF_HF_file)
LFHF_data = LFHF_data[1:11]
# display(LFHF_data)
merged = pd.merge(LFHF_data, result_df, on='trial_count')
merged.to_csv(data_dir + "/experimental_results.csv")
display(merged)
```
